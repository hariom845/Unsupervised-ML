# Principle Component Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
## Load dataset
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer_dataset.data, columns = cancer_dataset.feature_names)
```

```python
df.head()
```

```python
## Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

```python
scaled_data
```

```python
### Applying PCA algorithms
from sklearn.decomposition import PCA
```

```python
pca=PCA()
pca.fit_transform(scaled_data)
```

```python
pca.explained_variance_ #variance of all features
```

```python
pca = PCA(n_components=2) # extracting 2 features
```

```python
data_pca = pca.fit_transform(scaled_data) # out of all columns we have transformed in only two features
```

```python
data_pca
```

```python
## variance of feature one and two
pca.explained_variance_
```

```python
## what we can do with these 2 features
# lets visualize it

plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1], c=cancer_dataset['target'],cmap='plasma')
plt.xlabel('first principle component') # first pca is always greater than second one
plt.ylabel('second principle component')
# here datapoints are separable so we can perform any classification algorithm
```
