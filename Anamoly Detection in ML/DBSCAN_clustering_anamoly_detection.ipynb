# Anamoly detection using DBSCAN clusterring

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples = 750, factor=0.3, noise=0.1) # noise as outlier
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 0.10)
```

```python
dbscan.fit_predict(X)
dbscan.labels_ # -1 means it is outliers
```

```python
plt.scatter(X[:,0],X[:,1], c=dbscan.labels_)
```

```python
plt.scatter(X[:,0], X[:,1], c=y) # real points
```
